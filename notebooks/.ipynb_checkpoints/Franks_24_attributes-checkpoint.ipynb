```python
# Extract ALL indicators from World Bank Data
from pymongo import MongoClient
client = MongoClient('mongodb://mongo:27017/')
db = client['epidb']

WDI_indicators = db.WDI_indicators

for indicator in db.WDI_indicators.find():
    print(indicator["code"],"\t\t\t" ,indicator["name"])
```
